```python
import numpy as np
import scipy.io
import skimage.feature
import matplotlib
import matplotlib.pyplot as plt
 
data = scipy.io.loadmat('kfdata.mat')
noisy_data = data["data"]
true_data = data["true_data"]

```

```python
class KalmanFilter(object):
    
    def __init__(self, data):
        self.X_state_matrix = None
        self.P_process_matrix = None
        
        self.X_predicted_state_matrix = None
        self.P_predicted_process_matrix = None
        
        self.K_kalman_gain_matrix = None
        self.R_sensor_noise_matrix = None
        self.H_matrix = None
        self.I_identity_matrix = None
        
        self.U_control_matrix = None
        self.A_matrix = None
        self.B_matrix = None
        self.W_state_noise_matrix = None
        self.Q_process_noise_matrix = None
        
        self.Y_measurement_state_matrix = None
        self.Z_measurement_noise_matrix = None
        self.C_matrix = None
        return
    
    
    def Compute_Next_Prediction(self):
        self.Compute_Predicted_State()
        self.Compute_Predicted_VS_Measurement_State()
        self.Compute_Current_State()
        return self.X_current_state_matrix
    
    def Compute_Predicted_State(self):
        AX = numpy.dot(self.A_matrix, self.X_state_matrix)
        BU = numpy.dot(self.B_matrix, self.U_control_matrix)
        W = self.W_state_noise_matrix
        self.X_predicted_state_matrix = AX + BU + W
        
        AP = numpy.dot(self.A_matrix, self.P_process_matrix)
        APA_t = numpy.dot(AP, self.A_matrix.T)
        Q = self.Q_process_noise_matrix
        self.P_predicted_process_matrix = APA_t + Q
        return
    
    def Compute_Predicted_VS_Measurement_State(self):
        numerator_matrix = numpy.dot(self.P_predicted_process_matrix, self.H_matrix.T)
        HP_pred = numpy.dot(self.H_matrix, self.P_predicted_process_matrix)
        HPH_t = numpy.dot(HP_pred, self.H_matrix.T)
        R = self.R_sensor_noise_matrix
        denominator_matrix = HPH_t + R
        self.K_kalman_gain_matrix = numerator_matrix / denominator_matrix
        return 
    
    def Compute_Current_State():
        HX_pred = numpy.dot(self.H_matrix, self.X_predicted_state_matrix)
        S = self.Y_measurement_state_matrix - HX_pred
        KS = numpy.dot(self.K_kalman_gain_matrix, S)
        X_pred = self.X_predicted_state_matrix
        self.X_state_matrix = X_pred + KS
        
        KH = numpy.dot(self.K_kalman_gain_matrix, self.H_matrix)
        I = self.I_identity_matrix
        I_minus_KH = I - KH
        self.P_process_matrix = numpy.dot(I_minus_KH, self.X_predicted_state_matrix)
        return
        
        
        
        
```
