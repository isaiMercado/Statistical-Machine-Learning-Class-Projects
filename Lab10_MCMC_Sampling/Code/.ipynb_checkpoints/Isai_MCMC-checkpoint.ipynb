```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
```

```python
class MetropolisHastings:
    
    
    def __init__(self, times):
        self.times = times
        return
    
    
    # Our q is the easy distribution (In this case gaussian)
    def Q(self, x):
        return random.gauss(x, self.variance)
    
    
    # p is the hard distribution
    def P(self, x, t=1.0 ):
        return np.exp( -10*t*((x-2)**2) ) + 0.3*np.exp( -0.5*10*t*((x+1)**2) )

    
    def trueDistribution(self, x_domain):
        output = []
        for x in x_domain:
            output.append(self.P(x))
        return output
        
        
    def MCMC(self):
        
        state = 0.0
        states = [state]
        
        accepted_states  = 0

        for time in range(self.times - 1):
            
            state_current = states[time]  
            state_proposal = self.Q(state_current)
            
            a = (self.P(state_proposal) * self.Q(state_proposal)) / (self.P(state_current) * self.Q(state_current))
            
            r = min(1, a)
            
            u = random.random()
            
            if (u < r):
                states.append(state_proposal)
                accepted_states = accepted_states + 1
            else:
                states.append(state_current)
         
        print("Acceptance rate = " + str(accepted_states / self.times))
        return states
    
    
    def run(self, variance):
        
        self.variance = variance
        
        times_domain = range(self.times)
        mcmc_states1 = self.MCMC()
        mcmc_states2 = self.MCMC()
        mcmc_states3 = self.MCMC()
        
        figure = plt.figure()
        figure.set_size_inches(20, 10, forward = True)
        subplot = figure.add_subplot(111)
        subplot.plot(times_domain, mcmc_states1, color='red', label="States 1")
        subplot.plot(times_domain, mcmc_states2, color='blue', label="States 2")
        subplot.plot(times_domain, mcmc_states3, color='green', label="States 3")
        subplot.set_title("Metropolis Hastings MCMC States Over Time", fontsize=30)
        subplot.legend(loc=2,prop={'size':20})
        plt.show()
        
        x_domain = np.arange(-3, 3, 0.01)
        true_distribution = self.trueDistribution(x_domain)
        
        bins = 50
        sample_domain = np.linspace(-3, 3, bins)
        mcmc_histogram, f = np.histogram(mcmc_states1, bins = bins)
        
        print(str(mcmc_histogram))
        print(str(len(mcmc_histogram)))
        
        figure = plt.figure()
        figure.set_size_inches(20, 10, forward = True)
        subplot = figure.add_subplot(111)
        subplot.bar(sample_domain, mcmc_histogram, width=1/bins, color='gray', label="MCMC Samples")
        subplot.plot(x_domain, true_distribution, color='purple', label="True Distribution")
        subplot.set_title("Metropolis Hastings MCMC Distribution", fontsize=30)
        subplot.legend(loc=2,prop={'size':20})
        plt.show()


        
```

```python
mcmc1 = MetropolisHastings(times = 10000)
```

```python
mcmc1.run(variance = 0.1)
```

```python
mcmc1.run(variance = 1)
```

```python
mcmc1.run(variance = 10)
```
