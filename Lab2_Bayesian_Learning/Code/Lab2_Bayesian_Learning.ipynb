```python
import matplotlib.pyplot as plt
#import seaborn

# this allows you to debug
#import pdb  pdb.set_trace()

# this tells seaborn and matplotlib to generate plots inline in the notebook
%matplotlib inline  
 
# these two lines allow you to control the figure size
%pylab inline
pylab.rcParams['figure.figsize'] = (20.0, 4.0)


class Hypothesis(object):
   

    def __init__(self, userInput, gameObject, prior):
        self.userInput = userInput
        self.gameObject = gameObject
        self.currentPrior = prior
        self.currentPosterior = 0
        return
        
    
    def prior(self):
        return self.currentPrior
        
        
    def likelihood(self):
        output = 0.0
        if self.areMembers(self.userInput):
            memberProbability = ( 1 / self.hypothesisSpace() )
            output = memberProbability ** len(self.userInput)
        return output
    
    
    def hypothesisSpace(self):
        space = 0
        for inputNumber in range(self.gameObject.numberStart, self.gameObject.numberEnd):
            space = space + self.isMember(inputNumber)
        return space
    
    
    def posterior(self): 
        output = ( self.likelihood() * self.prior() ) / self.dataMargin()
        return output

    
    def dataMargin(self):
        summation = 0
        for hypothesis in self.gameObject.hypotheses:
            summation = summation + hypothesis.likelihood() * hypothesis.prior()
        return summation
    
    
    def isMember(self, inputNumber):
        pass
    
    
    def areMembers(self, numbersInput):
        members = True
        for numberInput in numbersInput:
            if (self.isMember(numberInput) == 0.0):
                members = False
        return members
    
    
    def posteriorPredictive(self, inputNumber):
        prediction = self.isMember(inputNumber) * self.posterior()
        return prediction
    
 
```

```python
DEBUG = False


class NumberGame(object):
    
    
    numberStart = 1;
    numberEnd = 101;
    
    
    def __init__(self, userInput):
        self.userInput = userInput
        self.hypotheses = []
        self.hypotheses.append(Even(userInput, self, 0.10))
        self.hypotheses.append(Odd(userInput, self, 0.10))
        self.hypotheses.append(Squares(userInput, self, 0.10))
        self.hypotheses.append(MultOf3(userInput, self, 0.02))
        self.hypotheses.append(MultOf4(userInput, self, 0.02))
        self.hypotheses.append(MultOf5(userInput, self, 0.02))
        self.hypotheses.append(MultOf6(userInput, self, 0.02))
        self.hypotheses.append(MultOf7(userInput, self, 0.02))
        self.hypotheses.append(MultOf8(userInput, self, 0.02))
        self.hypotheses.append(MultOf9(userInput, self, 0.02))
        self.hypotheses.append(MultOf10(userInput, self, 0.02))
        self.hypotheses.append(EndsIn0(userInput, self, 0.02))
        self.hypotheses.append(EndsIn1(userInput, self, 0.02))
        self.hypotheses.append(EndsIn2(userInput, self, 0.02))
        self.hypotheses.append(EndsIn3(userInput, self, 0.02))
        self.hypotheses.append(EndsIn4(userInput, self, 0.02))
        self.hypotheses.append(EndsIn5(userInput, self, 0.02))
        self.hypotheses.append(EndsIn6(userInput, self, 0.02))
        self.hypotheses.append(EndsIn7(userInput, self, 0.02))
        self.hypotheses.append(EndsIn8(userInput, self, 0.02))
        self.hypotheses.append(EndsIn9(userInput, self, 0.02))
        self.hypotheses.append(PowerOf2(userInput, self, 0.10))
        self.hypotheses.append(PowerOf3(userInput, self, 0.02))
        self.hypotheses.append(PowerOf4(userInput, self, 0.02))
        self.hypotheses.append(PowerOf5(userInput, self, 0.02))
        self.hypotheses.append(PowerOf6(userInput, self, 0.02))
        self.hypotheses.append(PowerOf7(userInput, self, 0.02))
        self.hypotheses.append(PowerOf8(userInput, self, 0.02))
        self.hypotheses.append(PowerOf9(userInput, self, 0.02))
        self.hypotheses.append(PowerOf10(userInput, self, 0.02))
        self.hypotheses.append(All(userInput, self, 0.02))
        self.hypotheses.append(PowerOf2And37(userInput, self, 0.02))
        self.hypotheses.append(PowerOf2AndNot32(userInput, self, 0.02))
        return
    
    
    def hypothesesNames(self):
        names = [hypothesis.__class__.__name__ for hypothesis in self.hypotheses]
        return names
            
       
    def normalizedPriors(self):
        priors = [hypothesis.prior() for hypothesis in self.hypotheses]
        priorsSummation = sum(priors)
        normalizedPriors = [prior / priorsSummation for prior in priors]
        if DEBUG: print "\nPriors " + str(normalizedPriors)
        return normalizedPriors
    
    
    def normalizedLikelihoods(self):
        normalizedlikelihoods = list()
        for hypothesis in self.hypotheses:
            if hypothesis.likelihood() != 0.0:
                normalizedlikelihoods.append(len(self.userInput) / hypothesis.hypothesisSpace())
            else:
                normalizedlikelihoods.append(0.0)
        if DEBUG: print "\nLikelihoods " + str(normalizedlikelihoods)
        return normalizedlikelihoods
    
    
    def normalizedPosteriors(self):
        posteriors = [hypothesis.posterior() for hypothesis in self.hypotheses]
        posteriorsSummation = sum(posteriors)
        normalizedPosteriors = [posterior / posteriorsSummation for posterior in posteriors]
        if DEBUG: print "\nPosteriors " + str(normalizedPosteriors)
        return normalizedPosteriors
    
    
    def normalizedPosteriorPredictives(self):
        posteriorPredictives = list()
        totalValidHypothesis = set()
        for inputNumber in range(self.numberStart, self.numberEnd):
            posteriorPredictives.append(0.0)
            for hypothesis in self.hypotheses:
                if hypothesis.posteriorPredictive(inputNumber) != 0.0:
                    posteriorPredictives[inputNumber - 1] = posteriorPredictives[inputNumber - 1] + 1.0
                    totalValidHypothesis.add(hypothesis)
        normalizedPosteriorPredictives = [posteriorPredictive / len(totalValidHypothesis) for posteriorPredictive in posteriorPredictives]  
        if DEBUG: print "\nPosterior Predictives " + str(normalizedPosteriorPredictives)
        return normalizedPosteriorPredictives
    
    
    def printPriorsTable(self):
        priors = self.normalizedPriors()
        self.plot(xDomain = range(len(priors)),
                  xValues = priors, 
                  xLabels = self.hypothesesNames(), 
                  yLabel  = "Probability", 
                  tittle  = "Priors",
                  yRange  = [0,1])
        return
    
    
    def printLikelihoodsTable(self):
        likelihoods = self.normalizedLikelihoods()    
        self.plot(xDomain = range(len(likelihoods)),
                  xValues = likelihoods, 
                  xLabels = self.hypothesesNames(), 
                  yLabel  = "Probability", 
                  tittle  = "Likelihoods",
                  yRange  = [0,1])
        return


    def printPosteriorsTable(self):
        posteriors = self.normalizedPosteriors()   
        self.plot(xDomain = range(len(posteriors)),
                  xValues = posteriors, 
                  xLabels = self.hypothesesNames(), 
                  yLabel  = "Probability", 
                  tittle  = "Posteriors",
                  yRange = [0,1])
        return


    def printPosteriorPredictiveTable(self):
        individualValidHypotheses = self.normalizedPosteriorPredictives()      
        self.plot(xDomain = range(len(individualValidHypotheses)),
                  xValues = individualValidHypotheses, 
                  xLabels = range(self.numberStart, self.numberEnd), 
                  yLabel  = "Probability", 
                  tittle  = "Posteriors Predictive",
                  yRange  = [0,1])
        return
    
    
    def printBestHypothesis(self):
        bestHypothesisPosterior = 0
        for hypothesis in self.hypotheses:
            if (hypothesis.posterior() > bestHypothesisPosterior):
                bestHypothesis = hypothesis
                bestHypothesisPosterior = bestHypothesis.posterior()
        print "\nBest Hypothesis: " + bestHypothesis.__class__.__name__
        return
    
    
    def plot(self, xDomain, xValues, xLabels, yLabel, tittle, yRange = None):    
        subplot = plt.subplot(111)
        subplot.bar(left = xDomain, 
                    height = xValues,
                    width=0.4,
                    color='b')
        subplot.set_ylabel(yLabel)
        subplot.set_title(tittle)
        subplot.set_xticks([x + 0.3 for x in xDomain])
        subplot.set_xticklabels(xLabels, rotation='vertical')
        if yRange is not None:
            subplot.set_ylim(yRange)
        plt.show()
        return
        
```
