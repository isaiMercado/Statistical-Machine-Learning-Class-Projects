```python
# Gaussian processes (GPs) extend multivariate Gaussian distributions to infinite dimensionality.
# Formally, a Gaussian process generates data located throughoutsome domain
# such that any finite subset of the range follows a multivariate Gaussian distribution.
# http://www.robots.ox.ac.uk/~mebden/reports/GPtutorial.pdf
class Kernel(object):


    #  |f |      / |0|   / K(X, X)  K(X, X∗) \ \
    #  |  | ~ N |  | |   |                    | |
    #  |f∗|      \ |0| , \ K(X∗, X) K(X∗, X∗) / / 
    # 
    # X* is the test point
    # X is the vector of known points
    #
    # K(X, X)   is a matrix 
    # K(X*, X)  is a row vector 
    # K(X*, X*) is a scalar
    # K(X, X*)  is a column vector
    def __init__(self, xData, yData, variance):
        self.xData = xData
        self.yData = yData
        self.variance = variance
        X_X_covMatrix = self.Kernel_X_X(xData)
        self.X_X_covMatrix_inv = numpy.linalg.pinv(X_X_covMatrix)
        return
    
    
    # This function returns the matrix of K(X,X)
    def Kernel_X_X(self, xData):
        X_X_covMatrix = numpy.empty(shape=[len(xData), len(xData)])
        for row in range(len(xData)):
            for col in range(len(xData)):
                X_X_covMatrix[row,col] = self.K(xData[row], xData[col]) + self.variance * self.D(row, col)
        return X_X_covMatrix
    
    
    # This function returns the row vector K(X*,X)
    def Kernel_xTest_X(self, xTest):
        xTest_X_covMatrix = numpy.empty(shape=[1, len(self.xData)])
        for col in range(len(self.xData)):
            xTest_X_covMatrix[0,col] = self.K(xTest, self.xData[col])
        return xTest_X_covMatrix
    
    
    # This function returns the column vector K(X, X*)
    def Kernel_X_xTest(self, xTest):
        X_xTest_covMatrix = numpy.empty(shape=[len(self.xData), 1])
        for row in range(len(self.xData)):
            X_xTest_covMatrix[row,0] = self.K(xTest, self.xData[row])
        return X_xTest_covMatrix
    
    
    # This function returns the scalar of K(X*,X*)
    def Kernel_xTest_xTest(self, xTest):
        i = len(self.xData) + 1
        j = len(self.xData) + 1
        xTest_xTest_variance = self.K(xTest, xTest) + self.variance * self.D(i, j)
        return xTest_xTest_variance
    
    
    # This function needs to be overwritten by the different Kernels (Which are a different class)
    def K(self, matrixX, matrixY):
        pass
    
    
    # The delta function returns a one if we are computing the kernel of a cell
    # in the diagonal of the matrix. This is done to simulate noise in the points that we know because 
    # measurement instrument are not perfect.
    #
    # | K(X,X)  K(X,X*) |
    # | K(X*,X) K(X*,X*)|
    #
    def D(self, i, j):
        if (i == j):
            return 1.0
        return 0.0
    
    
    # This function returns the possible mu of the test point
    # mu' = mu(X*) + K(X*,X) . K(X,X)^-1 . (Y - mu(X*)
    # where mu of X* is zero for simplicity, and . is the dot product
    def muPrime(self, xTest):
        mean = numpy.dot(self.Kernel_xTest_X(xTest), self.X_X_covMatrix_inv)
        mean = numpy.dot(mean, self.yData)
        mean = numpy.asscalar(mean[0])
        return mean
    
    
    # This function returns the possible variance of the test point
    # sigma' = K(X*,X*) - K(X*,X) . K(X,X)^-1 . K(X,X*)
    # . is the dot product
    def variancePrime(self, xTest):
        variance = numpy.dot(self.Kernel_xTest_X(xTest), self.X_X_covMatrix_inv)
        variance = numpy.dot(variance, self.Kernel_X_xTest(xTest))
        variance = self.Kernel_xTest_xTest(xTest) - variance
        variance = numpy.asscalar(variance[0])
        return variance  
    
   
    # This function is in charge of making a subspace of points that will be used as x-values in the plot
    # and this x-values will be the test points of out muPrime, and variancePrime functions to plot the 
    # possible distribution of the x-values.
    # To make it easier for the grader the graph starts at -2, and goes until 10
    def Graph_Regression(self):
        regressionXValues = numpy.arange(-2, 10, .1)
        regressionMeans = numpy.arange(-2, 10, .1)
        regressionVarianceTop = numpy.arange(-2, 10, .1)
        regressionVarianceBottom = numpy.arange(-2, 10, .1)
        
        index = 0
        for xTest in regressionXValues: 
            mean = self.muPrime(xTest)
            variance = self.variancePrime(xTest)
            varianceTop = mean + variance
            varianceBottom = mean - variance
            regressionVarianceTop[index] = varianceTop
            regressionMeans[index] = mean
            regressionVarianceBottom[index] = varianceBottom
            index = index + 1
            
        self.plot(dataXValues = self.xData,
                  dataYValues = self.yData, 
                  regressionXValues = regressionXValues, 
                  regressionMeans = regressionMeans,
                  regressionVarianceTop = regressionVarianceTop,
                  regressionVarianceBottom = regressionVarianceBottom,
                  tittle = "Regression Graph",
                  yBoundaries = [-3, 3],
                  xBoundaries = [-3, 11])
        return
        

    # This function receives the information made by the Graph_Regression function 
    # and displays the plot
    def plot(self, dataXValues, dataYValues, regressionXValues, regressionMeans, regressionVarianceTop, 
             regressionVarianceBottom, tittle, xLabels = None, yLabel = None, xBoundaries = None, yBoundaries = None): 
        
        figure = pyplot.figure()
        figure.set_size_inches(20, 10, forward = True)
        subplot = figure.add_subplot(111)
        subplot.fill_between(regressionXValues, 
                             regressionVarianceTop, 
                             regressionVarianceBottom, 
                             facecolor='lightskyblue',
                             alpha=0.5)
        subplot.plot(regressionXValues, regressionMeans, color='purple', linewidth=4, alpha=0.5)
        subplot.scatter(dataXValues, dataYValues, color='red')
        subplot.set_title(tittle)
        
        if yLabel is not None:
            subplot.set_ylabel(yLabel)
    
        if xLabels is not None:
            subplot.set_xticks([x + 0.3 for x in dataXValues])
            subplot.set_xticklabels(xLabels, rotation='vertical')
        
        if yBoundaries is not None:
            subplot.set_ylim(yBoundaries)
            
        if xBoundaries is not None:
            subplot.set_xlim(xBoundaries)
            
        pyplot.show()
        return

    
```

```python
class LinearKernel(Kernel):
    def K(self, point_i, point_j):
        kernel = point_i * point_j 
        return kernel
```
